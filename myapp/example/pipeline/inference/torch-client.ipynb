```python
# 弹性伸缩测试

import requests
from concurrent.futures import ThreadPoolExecutor

# 定义要启动的线程数（即客户端数）
num_clients = 100
request_num = 10000
try_num=0
def load_test():
    global try_num
    SERVER_URL = 'http://resnet50-202208012.service:8080/predictions/resnet50'
    IMAGE_PATH = 'smallcat.jpg'
    files = {'data': open(IMAGE_PATH, 'rb')}
    response = requests.post(SERVER_URL, files=files)
    # 打印响应状态码
    print(try_num,'成功' if response.status_code==200 else '失败',flush=True)
    try_num+=1

# 创建一个ThreadPoolExecutor实例
with ThreadPoolExecutor(max_workers=num_clients) as executor:
    # 使用executor提交load_test任务，参数为url
    for _ in range(request_num):
        executor.submit(load_test)
```
